```
# Unit tests
import unittest

class TestStringRegexIO(unittest.TestCase):
    def test_count_vowels(self):
        self.assertEqual(count_vowels("hello"), 2)
        self.assertEqual(count_vowels("Python"), 1)

    def test_is_valid_email(self):
        self.assertTrue(is_valid_email("test@example.com"))
        self.assertFalse(is_valid_email("invalid_email"))

    def test_extract_urls(self):
        text = "Visit https://www.example.com and http://test.com"
        self.assertEqual(extract_urls(text), ["https://www.example.com", "http://test.com"])

    def test_replace_in_file(self):
        with open("test_file.txt", "w") as f:
            f.write("Hello world")
        replace_in_file("test_file.txt", "world", "Python")
        with open("test_file.txt", "r") as f:
            self.assertEqual(f.read(), "Hello Python")
        os.remove("test_file.txt")

    def test_find_palindromes(self):
        self.assertEqual(find_palindromes("level radar python"), ["level", "radar"])

    def test_remove_non_alphanumeric(self):
        self.assertEqual(remove_non_alphanumeric("Hello, World! 123"), "HelloWorld123")

    def test_read_csv(self):
        with open("test.csv", "w") as f:
            f.write("name,age\nAlice,30\nBob,25")
        result = read_csv("test.csv")
        self.assertEqual(result, [{"name": "Alice", "age": "30"}, {"name": "Bob", "age": "25"}])
        os.remove("test.csv")

    def test_find_phone_numbers(self):
        text = "Call me at 123-456-7890 or (987) 654-3210"
        self.assertEqual(find_phone_numbers(text), ["123-456-7890", "(987) 654-3210"])

    def test_capitalize_words(self):
        self.assertEqual(capitalize_words("hello world"), "Hello World")

    def test_is_valid_ipv4(self):
        self.assertTrue(is_valid_ipv4("192.168.0.1"))
        self.assertFalse(is_valid_ipv4("256.1.2.3"))

    def test_pretty_print_json(self):
        with open("test.json", "w") as f:
            f.write('{"name":"John","age":30}')
        pretty_print_json("test.json")
        os.remove("test.json")

    def test_extract_dates(self):
        text = "Dates: 01/01/2022 and 31/12/2023"
        self.assertEqual(extract_dates(text), ["01/01/2022", "31/12/2023"])

    def test_word_frequency(self):
        with open("test_words.txt", "w") as f:
            f.write("apple banana apple cherry")
        self.assertEqual(word_frequency("test_words.txt"), {"apple": 2, "banana": 1, "cherry": 1})
        os.remove("test_words.txt")

    def test_is_valid_password(self):
        self.assertTrue(is_valid_password("P@ssw0rd"))
        self.assertFalse(is_valid_password("weakpass"))

    def test_find_longest_word(self):
        with open("test_longest.txt", "w") as f:
            f.write("Python programming language")
        self.assertEqual(find_longest_word("test_longest.txt"), "programming")
        os.remove("test_longest.txt")
```

```
# Unit tests
import unittest

class TestDataStructures(unittest.TestCase):
    def test_second_largest(self):
        self.assertEqual(second_largest([1, 3, 4, 5, 0, 2]), 4)
        self.assertEqual(second_largest([1, 1, 1, 1, 1]), 1)

    def test_merge_sorted_lists(self):
        self.assertEqual(merge_sorted_lists([1, 3, 5], [2, 4, 6]), [1, 2, 3, 4, 5, 6])
        self.assertEqual(merge_sorted_lists([1, 2, 3], []), [1, 2, 3])

    def test_remove_duplicates(self):
        self.assertEqual(remove_duplicates([1, 2, 2, 3, 4, 3, 5]), [1, 2, 3, 4, 5])
        self.assertEqual(remove_duplicates([1, 1, 1, 1]), [1])

    def test_list_intersection(self):
        self.assertEqual(list_intersection([1, 2, 3, 4], [3, 4, 5, 6]), [3, 4])
        self.assertEqual(list_intersection([1, 2, 3], [4, 5, 6]), [])

    def test_element_frequency(self):
        self.assertEqual(element_frequency([1, 2, 2, 3, 3, 3]), {1: 1, 2: 2, 3: 3})
        self.assertEqual(element_frequency([]), {})

    def test_invert_dict(self):
        self.assertEqual(invert_dict({'a': 1, 'b': 2, 'c': 3}), {1: 'a', 2: 'b', 3: 'c'})
        self.assertEqual(invert_dict({}), {})

    def test_max_value_key(self):
        self.assertEqual(max_value_key({'a': 5, 'b': 2, 'c': 8}), 'c')
        self.assertIsNone(max_value_key({}))

    def test_merge_dicts(self):
        self.assertEqual(merge_dicts({'a': 1, 'b': 2}, {'c': 3, 'd': 4}), {'a': 1, 'b': 2, 'c': 3, 'd': 4})
        self.assertEqual(merge_dicts({'a': 1}, {'a': 2}), {'a': 2})

    def test_symmetric_difference(self):
        self.assertEqual(symmetric_difference({1, 2, 3}, {3, 4, 5}), {1, 2, 4, 5})
        self.assertEqual(symmetric_difference({1, 2}, {3, 4}), {1, 2, 3, 4})

    def test_is_subset(self):
        self.assertTrue(is_subset({1, 2}, {1, 2, 3, 4}))
        self.assertFalse(is_subset({1, 5}, {1, 2, 3, 4}))

    def test_create_matrix(self):
        self.assertEqual(create_matrix(2, 3), [[0, 0, 0], [0, 0, 0]])
        self.assertEqual(create_matrix(1, 1), [[0]])

    def test_transpose_matrix(self):
        self.assertEqual(transpose_matrix([[1, 2], [3, 4]]), [[1, 3], [2, 4]])
        self.assertEqual(transpose_matrix([[1, 2, 3]]), [[1], [2], [3]])

    def test_add_matrices(self):
        self.assertEqual(add_matrices([[1, 2], [3, 4]], [[5, 6], [7, 8]]), [[6, 8], [10, 12]])
        self.assertEqual(add_matrices([[1]], [[2]]), [[3]])

    def test_multiply_matrices(self):
        self.assertEqual(multiply_matrices([[1, 2], [3, 4]], [[5, 6], [7, 8]]), [[19, 22], [43, 50]])
        self.assertEqual(multiply_matrices([[1, 2]], [[3], [4]]), [[11]])

    def test_max_per_row(self):
        self.assertEqual(max_per_row([[1, 2, 3], [4, 5, 6], [7, 8, 9]]), [3, 6, 9])
        self.assertEqual(max_per_row([[1], [2], [3]]), [1, 2, 3])
```

```
import unittest

class TestClassesAndModules(unittest.TestCase):
    def test_rectangle(self):
        r = Rectangle(4, 5)
        self.assertEqual(r.area(), 20)
        self.assertEqual(r.perimeter(), 18)

    def test_circle(self):
        c = Circle(3)
        self.assertAlmostEqual(c.area(), 28.27, places=2)
        self.assertAlmostEqual(c.circumference(), 18.85, places=2)

    def test_bank_account(self):
        account = BankAccount()
        account.deposit(100)
        account.withdraw(30)
        self.assertEqual(account.check_balance(), 70)

    def test_person(self):
        p = Person("Alice", 30)
        self.assertEqual(p.introduce(), "My name is Alice and I am 30 years old.")

    def test_employee(self):
        e = Employee("Bob", 35, 50000)
        self.assertEqual(e.introduce(), "My name is Bob and I am 35 years old. My salary is $50000.")

    def test_car(self):
        car = Car("Toyota", "Corolla", 2020)
        self.assertEqual(car.display_info(), "2020 Toyota Corolla")

    def test_stack(self):
        s = Stack()
        s.push(1)
        s.push(2)
        self.assertEqual(s.pop(), 2)
        self.assertFalse(s.is_empty())

    def test_queue(self):
        q = Queue()
        q.enqueue(1)
        q.enqueue(2)
        self.assertEqual(q.dequeue(), 1)
        self.assertFalse(q.is_empty())

    def test_point(self):
        p1 = Point(0, 0)
        p2 = Point(3, 4)
        self.assertEqual(p1.distance(p2), 5)

    def test_book(self):
        book = Book("Python Basics", "John Doe", 29.99)
        self.assertEqual(book.display_info(), "Python Basics by John Doe, Price: $29.99")

    def test_calculator(self):
        self.assertEqual(add(2, 3), 5)
        self.assertEqual(subtract(5, 3), 2)
        self.assertEqual(multiply(2, 4), 8)
        self.assertEqual(divide(10, 2), 5)

    def test_temperature(self):
        temp = Temperature()
        self.assertEqual(temp.celsius_to_fahrenheit(0), 32)
        self.assertEqual(temp.fahrenheit_to_celsius(32), 0)

    def test_student(self):
        student = Student("Eve")
        student.add_grade(85)
        student.add_grade(90)
        self.assertEqual(student.average_grade(), 87.5)

    def test_timer(self):
        timer = Timer()
        timer.start()
        time.sleep(0.1)
        timer.stop()
        self.assertGreater(timer.elapsed_time(), 0.1)

    def test_string_utils(self):
        self.assertEqual(string_utils.reverse_string("hello"), "olleh")
        self.assertTrue(string_utils.is_palindrome("racecar"))
        self.assertFalse(string_utils.is_palindrome("hello"))

if __name__ == '__main__':
    unittest.main()
```
